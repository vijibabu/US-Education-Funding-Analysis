## 2. Exploring relationships

### Correlations

```python
# Create correlation matrix

df_state.corr()
```

```python
state_sub = df_state[['enroll', 'total_rev','fed_rev', 'st_rev', 'loc_rev','total_exp', 'instruction_exp', 'service_exp', 'other_exp', 'capital_exp','avg_score']] 
```

```python
sns.set_style({'axes.facecolor': 'black', 'figure.facecolor':'black','text.color':'white', 'xtick.color': 'white', 'ytick.color':'white', 'axes.edgecolor': 'white','xlabel.color':'white','ylabel.color':'white'})
```

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Create the correlation heatmap 
corr = sns.heatmap(state_sub.corr(), annot = True, ax = ax) 
plt.savefig("corrmatrix.png")
```

### There is a very strong positive correlation between enroll, revenue and expenditure. As the count of enroll increases the total revenue received and expediture incured increases. But anyway there is no significant relationship between the average score obtained in the test and funding received and count of enroll.

## Scatterplots

```python
# Create a scatterplot to show the relationship between revenue and expenditure
sns.lmplot(x = 'total_rev', y ='total_exp' , data = df_state, line_kws = {'color': 'red'})

plt.title('Relationship between revenue and expenditure', fontweight = 'bold');
plt.xlabel('Revenue', fontweight = 'bold', color = 'white');
plt.ylabel ('Expenditure', fontweight = 'bold', color = 'white');
plt.ticklabel_format(axis = 'y', style = '', scilimits = (0,10));
plt.ticklabel_format(axis = 'x', style = '', scilimits = (0,10));
plt.savefig("Scatter-rev-exp")
```

### There is strong correlation between the funding received and the expenditure made. From the plot we could see that all the data points are closely arranged forming a linear positive relationship. This means that as the revenue increases the expenditure increases.

```python
# Create a scatterplot for enroll and revenue
sns.lmplot(x = 'enroll', y = 'total_rev', data = df_state, line_kws = {'color': 'red'})
plt.title('Relationship between revenue and enroll', fontweight = 'bold');
plt.xlabel('Enroll', fontweight = 'bold', color = 'white');
plt.ylabel ('Revenue', fontweight = 'bold', color = 'white');
plt.ticklabel_format(axis = 'y', style = '', scilimits = (0,10));
plt.ticklabel_format(axis = 'x', style = '', scilimits = (0,10));
plt.savefig("Scatter-rev-enroll")
```

### There is a positive relationship between enroll and revenue. As the enroll increases the revenue increases. 

```python
# Create a scatterplot for avg_score and revenue

sns.lmplot(x = 'avg_score', y = 'total_rev', data = df_state, line_kws = {'color': 'red'})
plt.title('Relationship between revenue and average score', fontweight = 'bold');
plt.xlabel('Average Score', fontweight = 'bold', color = 'white');
plt.ylabel ('Revenue', fontweight = 'bold', color = 'white');
plt.ticklabel_format(axis = 'y', style = '', scilimits = (0,10));
plt.ticklabel_format(axis = 'x', style = '', scilimits = (0,10));
plt.savefig("Scatter-rev-score")
```

### There is no significant relationship between the score and revenue. The data is widely spread across.

## Categorical plot

```python
# Creating histogram to find the distribution

sns.histplot(df_state['enroll'], bins = 20, kde = True);
plt.title('Distribution of enroll', fontweight = 'bold');
plt.xlabel('Enroll',fontweight = 'bold', color = 'white');
plt.ylabel ('Count', fontweight = 'bold', color = 'white');
plt.ticklabel_format(axis = 'y', style = '', scilimits = (0,10));
plt.ticklabel_format(axis = 'x', style = '', scilimits = (0,10));
plt.savefig("Hist-enroll")
```

```python
df_state.loc[df_state['enroll'] <= 1000000, 'enroll_category'] = 'Low'
```

```python
df_state.loc[(df_state['enroll'] > 1000000) & (df_state['enroll'] <= 3000000), 'enroll_category'] = 'Medium'
```

```python
df_state.loc[df_state['enroll'] > 3000000, 'enroll_category'] = 'High'
```

```python
df_state['enroll_category'].value_counts()
```

```python
# Create a categorical plot 

sns.set(style="ticks")
g2 = sns.catplot(x="state", y="total_rev", hue="enroll_category", data=df_state, height = 5, aspect = 2)
plt.title('Statewise distribution of revenue and enroll category', fontweight = 'bold');
plt.xlabel('State',fontweight = 'bold');
plt.xticks(rotation = 90)
plt.ylabel ('Revenue', fontweight = 'bold');
plt.ticklabel_format(axis = 'y', style = '', scilimits = (0,10));
plt.savefig("categorical-enroll")
```

```python
# Create a categorical plot 

sns.set(style="ticks")
g3 = sns.catplot(x="state", y="total_exp", hue="enroll_category", data=df_state, height = 5, aspect = 2)
plt.title('Statewise distribution of expenditure and enroll category', fontweight = 'bold');
plt.xlabel('State',fontweight = 'bold');
plt.xticks(rotation = 90)
plt.ylabel ('Expenditure', fontweight = 'bold');
plt.ticklabel_format(axis = 'y', style = '', scilimits = (0,10));
plt.savefig("categorical-enroll-exp")
```

### The states with high count of enroll receives the high revenue and also incurs the high expenses. There are few states that has low enroll count but receives same revenue and incurs same expenditure as the medium count enroll.
